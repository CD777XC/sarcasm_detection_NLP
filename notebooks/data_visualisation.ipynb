```python
# Data modules
import pandas as pd
import numpy as np

# Visualisation tools
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../clean_data/sarcasm_headlines_v2_clean.csv')
```

```python
df.head()
```

```python
sequences = df['cleaned_headline']
count = sequences.str.split().str.len()
```

```python
count
```

```python
# Plotting sequences length distribution
plt.rcParams["figure.figsize"] = (12,8) # Define figure size
ax = sns.histplot(count, color='orangered', kde=True, binwidth=2, label='Number of words') # Instanciating the histplot

# Setting background color
ax.set_facecolor("bisque") 
plt.gcf().set_facecolor("bisque")

# Defining axis labels and sizes
plt.xlabel('Sequences words count') 
plt.ylabel('Distribution density')
plt.xlim(0, 100)

# Defining plot title and settings
plt.title('Sentences Length Distribution')
plt.grid()
plt.legend()

# Saving the plot
plt.savefig('word_distribution', bbox_inches='tight', pad_inches = 0)
plt.show()
```
